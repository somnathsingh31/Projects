```
import yfinance as yf
start_date = '2006-01-01'
end_date = '2016-01-01'
# Add multiple space separated tickers here
#ticker = 'GOOGL MSFT TSLA'
BAC = yf.download("BAC GOOGL MSFT TSLA", start_date, end_date)
BAC.head()
```

```
ticker = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```
BAC = yf.download(ticker[0], start_date, end_date)
C = yf.download(ticker[1], start_date, end_date)
GS = yf.download(ticker[2], start_date, end_date)
JPM = yf.download(ticker[3], start_date, end_date)
MS = yf.download(ticker[4], start_date, end_date)
WFC = yf.download(ticker[5], start_date, end_date)
```

```
WFC.head()
```

```
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis = 1, keys = ticker)
bank_stocks.head()
```

```
bank_stocks.columns.names = (['Bank Ticker', 'Stock Info'])
bank_stocks.head()
```

```
bank_stocks.xs('Close', axis = 1, level = 'Stock Info').max()
```

```
returns = pd.DataFrame()
```

```
for tick in ticker:
  returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```
